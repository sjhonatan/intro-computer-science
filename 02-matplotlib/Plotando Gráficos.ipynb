```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(9)
y = np.arange(9)
```

### Gráfico sobre plantação de batatas

```python
plt.scatter(x,y)
plt.grid(True)
plt.axes().set_aspect('equal','datalim')
plt.title('Crescimento de batatas')
plt.show()
```

```python
plt.scatter(x,2*y)
plt.grid(True)
plt.axes().set_aspect('equal','datalim')
plt.title('Crescimento de repolhos')
plt.show()
```
